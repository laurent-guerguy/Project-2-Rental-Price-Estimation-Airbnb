### Data cleaning

```python
# Check the types of the columns
listings.dtypes
```

```python
# Check the Nan's in all columns of the listings dataframe
listings.isna().sum(axis = 0)
```

There are a few columns with the wrong type so they need to be modified.

First I want to change the values in host_response_rate from % to a ratio number and change the type to a float

```python
# In order to be able to remove the % symbol using the replace method I need to replace the Nan in host_response_rate by an 
# empty string. Failure to do so will prevent me to use the replace method as it says "replace method does not work with float"
# because Nan is considered as a float.
listings[['host_response_rate']] = listings[['host_response_rate']].fillna("")
```

```python
listings.head()
```

```python
# Check if the Nans in host_response_rate have been replaced
listings.isna().sum(axis = 0)
```

```python
# Remove % from the host_response_ratio
listings["host_response_rate"] = listings["host_response_rate"].apply(lambda x: x.replace("%", ""))
```

```python
# Check it worked
listings.head()
```

```python
# Change the column type to float
listings['host_response_rate'] = pd.to_numeric(listings['host_response_rate'], errors='coerce')
```

```python
# Check the change of type has worked
listings.dtypes
```

```python
# Convert the numbers to the real ratio by dividing them by 100
listings["host_response_rate"] = listings["host_response_rate"].apply(lambda x: x/100)
```

```python
# Check the operation worked
listings.head()
```

```python
# Remove the $ symbol from all the prices columns.
listings = listings.replace({'\$': ''}, regex=True)
```

```python
# Check it worked
listings.head()
```

```python
# Replace the Nans with an empty string
listings[["price", "weekly_price", "monthly_price", "security_deposit", "cleaning_fee", "extra_people"]] = listings[["price", "weekly_price", "monthly_price", "security_deposit", "cleaning_fee", "extra_people"]].fillna("")
```

```python
# Convert all the prices to float
cols = ["price", "weekly_price", "monthly_price", "security_deposit", "cleaning_fee", "extra_people"]
listings[cols] = listings[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
# Check the types of columns
listings.dtypes
```

```python
# Save the listings cleaned dataframe as a csv called listings
listings.to_csv("datasets/listings_cleaned.csv")
```
